```python
import pandas as pd
import matplotlib.pyplot as plt

# Leitura do dataset paises.csv
df_paises = pd.read_csv('C:\\Users\\Defen\\Downloads\\paises.csv', sep=';')

```

```python
# Filtrar países da América do Norte
df_na = df_paises[df_paises['Region'].str.contains('America', case=False, na=False)]





# Plotar as curvas de Birthrate e Deathrate
plt.figure(figsize=(10, 6))
plt.plot(df_na['Country'], df_na['Birthrate'], label='Taxa de Natalidade (Birthrate)', marker='o')
plt.plot(df_na['Country'], df_na['Deathrate'], label='Taxa de Mortalidade (Deathrate)', marker='x')
plt.xticks(rotation=45)
plt.ylabel('Taxa')
plt.title('Taxa de Natalidade e Mortalidade - América do Norte')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Leitura do dataset space.csv
df  = pd.read_csv('C:\\Users\\Defen\\Downloads\\paises.csv', sep=';')


# Filtra apenas países da América do Norte
df_na = df[df['Region'].str.contains("NORTHERN AMERICA", case=False, na=False)]

# Remove valores ausentes e converte a taxa de nascimento para float
df_na['Birthrate'] = pd.to_numeric(df_na['Birthrate'], errors='coerce')
df_na = df_na.dropna(subset=['Birthrate'])

# Gráfico
plt.figure(figsize=(10, 5))
plt.bar(df_na['Country'], df_na['Birthrate'], color='skyblue')
plt.title('Taxa de Nascimento - América do Norte')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Taxa de Nascimento')
plt.tight_layout()
plt.show()

```

```python
# Converte taxas para float
df_na['Deathrate'] = pd.to_numeric(df_na['Deathrate'], errors='coerce')
df_na = df_na.dropna(subset=['Birthrate', 'Deathrate'])

# Gráfico de dispersão
plt.figure(figsize=(8, 5))
plt.scatter(df_na['Birthrate'], df_na['Deathrate'], color='orange')

# Adiciona nomes dos países nos pontos
for i, row in df_na.iterrows():
    plt.text(row['Birthrate'], row['Deathrate'], row['Country'], fontsize=8)

plt.title('Taxa de Nascimento vs. Taxa de Mortalidade - América do Norte')
plt.xlabel('Taxa de Nascimento')
plt.ylabel('Taxa de Mortalidade')
plt.grid(True)
plt.tight_layout()
plt.show()
```
